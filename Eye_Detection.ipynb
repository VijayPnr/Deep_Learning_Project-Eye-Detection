#Eye Detection

```python
import cv2
```

```python
eye_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/Data Science Luminar/Deep Learning (Dec-Jan)/DL Project/haarcascade_eye.xml')
```

```python
image = cv2.imread('/content/drive/MyDrive/Data Science Luminar/Deep Learning (Dec-Jan)/DL Project/mirror11.jpg')

gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)                                    #convert to gray scale
```

```python
eyes = eye_cascade.detectMultiScale(gray,scaleFactor=1.3,minNeighbors=5,minSize=(30,30))          #detect eye from gray scale
```

```python
print("Found {0} eyes!".format(len(eyes)))                                      #count of eyes
for (x,y,w,h) in eyes:                                                          #4 parameters , x,y-> axis , w,h-> width & height of rectangle
  cv2.rectangle(image,(x,y),(x+w, y+h),(0,255,0),3)                             #color-Green, thickness-3
```
